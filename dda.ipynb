```python
import matplotlib.pyplot as plt

def dda(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1
    steps = int(max(abs(dx), abs(dy)))
    x_inc = dx / steps
    y_inc = dy / steps
    x, y = x1, y1
    points_x = []
    points_y = []
    for _ in range(steps + 1):
        points_x.append(round(x))
        points_y.append(round(y))
        x += x_inc
        y += y_inc
    return points_x, points_y

x1 = float(input("Enter the value for x1: "))
y1 = float(input("Enter the value for y1: "))
x2 = float(input("Enter the value for x2: "))
y2 = float(input("Enter the value for y2: "))

px, py = dda(x1, y1, x2, y2)

plt.plot(px, py, marker='o')
plt.title("Line using DDA Algorithm")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```
